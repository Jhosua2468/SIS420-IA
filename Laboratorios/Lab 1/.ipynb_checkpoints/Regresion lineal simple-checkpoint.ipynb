# Importacion de librerias y carga del dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("insurance.csv")
print(df.head())
```

# Seleccion de X y y

```python
X = df["bmi"].values
y = df["charges"].values

# 4. Visualizar datos antes del entrenamiento
plt.scatter(X, y, c="blue", alpha=0.5)
plt.xlabel("BMI")
plt.ylabel("Charges")
plt.title("Datos iniciales: BMI vs Charges")
plt.show()
```

```python
m = y.size 

# Agregar columna de 1s manualmente (intercepto)
X = np.stack([np.ones(m), X], axis=1)

print(y)
```

# Funcin de costo

```python
def calcularCosto(X, y, theta):
    m = y.size
    J = (1/(2*m)) * np.sum(np.square(np.dot(X, theta) - y))
    return J
```

# Descenso por el gradiente

```python
def gradientDescent(X, y, theta, alpha, num_iters):
    m = y.shape[0]
    theta = theta.copy()
    J_history = []

    for i in range(num_iters):
        theta = theta - (alpha/m) * (np.dot(X, theta) - y).dot(X)
        J_history.append(calcularCosto(X, y, theta))

    return theta, J_history
```

# Entrenar modelo

```python
# Hiperparametros
theta_inicial = np.zeros(2)   # intercepto y pendiente
alpha = 0.00001 
num_iters = 500


theta_final, J_history = gradientDescent(X, y, theta_inicial, alpha, num_iters)

print("Theta final (intercepto, pendiente):", theta_final)
```

# Visualizar convergencia del costo

```python
plt.plot(range(len(J_history)), J_history)
plt.xlabel("Iteraciones")
plt.ylabel("Costo (J)")
plt.title("Convergencia del gradiente descendente")
plt.show()
```

```python
# 8. Visualizar recta ajustada
plt.scatter(df["bmi"], df["charges"], c="blue", label="Datos reales")

# Predicciones con la recta ajustada
x_vals = np.linspace(df["bmi"].min(), df["bmi"].max(), 100)
X_pred = np.stack([np.ones(x_vals.shape[0]), x_vals], axis=1)
y_pred = np.dot(X_pred, theta_final)

plt.plot(x_vals, y_pred, c="red", label="Recta ajustada")
plt.xlabel("BMI")
plt.ylabel("Charges")
plt.legend()
plt.show()
```
